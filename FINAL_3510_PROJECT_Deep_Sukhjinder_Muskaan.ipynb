## Feature engineering
Adding new features to above pivot table by creating lags for previous three years for each company by release.

#**PREDICTION FOR CARBON**

```python
def predict_carbon_monoxide(oil_price_change_percent):
    """
    Predicts carbon monoxide releases based on oil price changes.

    Args:
      oil_price_change_percent: Percentage change in oil price (positive for increase, negative for decrease).

    Returns:
      Predicted category of carbon monoxide release ('Low Release', 'Medium Release', 'High Release').
    """
    last_year_release = df_with_lags["Release_t-1"].iloc[-1]
    two_year_ago_release = df_with_lags["Release_t-2"].iloc[-1]
    three_year_ago_release = df_with_lags["Release_t-3"].iloc[-1]
    num_employees = df_with_lags['Number of employees_x'].iloc[-1]
    year = 2023 # Prediction year
    current_oil_price = df_with_lags["Average amount for oil price_y"].iloc[-1]
    new_oil_price = current_oil_price * (1+oil_price_change_percent/100)

    input_data = pd.DataFrame({
            "Release_t-1": [last_year_release],
            "Release_t-2": [two_year_ago_release],
            "Release_t-3": [three_year_ago_release],
            "Number of employees_x": [num_employees],
            "Year": [year],
            "Average amount for oil price_y": [new_oil_price]
        })

    # Make the prediction
    predicted_category = rf_classifier.predict(input_data)[0]
    return predicted_category

oil_price_increase = predict_carbon_monoxide(10) # 10% increase in oil price
oil_price_decrease = predict_carbon_monoxide(-10) # 10% decrease in oil price

print(f"Predicted release category with 10% increase in oil price: {oil_price_increase}")
print(f"Predicted release category with 10% decrease in oil price: {oil_price_decrease}")
```
